```python
X = pitchers[features_2]
y = pitchers['TARGET']

print(X.shape)
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.33, random_state=42)
```

```python
y_train.value_counts(normalize=True)
```

```python
ss = StandardScaler()
ss.fit(X_train)
X_train_sc = ss.transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
# linear Regression
lr = LinearRegression()
lr.fit(X_train_sc, y_train)
lr_preds = lr.predict(X_test_sc)
print(f' Linear Regression Train Score: {lr.score(X_train_sc, y_train)}')
print(f' Linear Regression Test Score: {lr.score(X_test_sc, y_test)}')
```

```python
# KNN Regression
knn = KNeighborsRegressor()
knn.fit(X_train_sc, y_train)
knn_preds = knn.predict(X_test_sc)
print(f' KNN Train Score: {knn.score(X_train_sc, y_train)}')
print(f' KNN Test Score: {knn.score(X_test_sc, y_test)}')
```

```python
# Decision Tree Regression
dtr = DecisionTreeRegressor()
dtr.fit(X_train_sc, y_train)
dtr_preds = dtr.predict(X_test_sc)
print(f' Decision Tree Train Score: {dtr.score(X_train_sc, y_train)}')
print(f' Decision Tree Test Score: {dtr.score(X_test_sc, y_test)}')
```

```python
# Bagging Regressor
bag = BaggingRegressor()
bag.fit(X_train_sc, y_train)
bag_preds = bag.predict(X_test_sc)
print(f' Bagging Regressor Train Score: {bag.score(X_train_sc, y_train)}')
print(f' Bagging Regressor Test Score: {bag.score(X_test_sc, y_test)}')
```

```python
# Random Forest Regressor
rfr = RandomForestRegressor()
rfr.fit(X_train_sc, y_train)
rfr_preds = rfr.predict(X_test_sc)
print(f' Random Forest Regressor Train Score: {rfr.score(X_train_sc, y_train)}')
print(f' Random Forest Regressor Test Score: {rfr.score(X_test_sc, y_test)}')
```

```python
# AdaBoost Regressor
ada = AdaBoostRegressor()
ada.fit(X_train_sc, y_train)
ada_preds = ada.predict(X_test_sc)
print(f' AdaBoost Train Score: {ada.score(X_train_sc, y_train)}')
print(f' AdaBoost Test Score: {ada.score(X_test_sc, y_test)}')
```

```python
# SVR 
svr = SVR()
svr.fit(X_train_sc, y_train)
svr_preds = svr.predict(X_test_sc)
print(f' SVR Train Score: {svr.score(X_train_sc, y_train)}')
print(f' SVR Test Score: {svr.score(X_test_sc, y_test)}')
```
